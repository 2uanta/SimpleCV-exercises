# Hello World

```python
from SimpleCV import Camera, Display
import time
# Initialize the camera
cam = Camera()
# Initialize the display
display = Display()
# Snap a picture using the camera
img = cam.getImage()
# Show some text
img.drawText("Hello World!") # this will start display on screen
# Show the picture on an external window
img.save(display) # and screen is refreshed with the camera snapshot
time.sleep(5) # wait 5 seconds before closing the display
display.quit()
```

```python
# display same image in this notebook
img.save("show-to-display.jpg")
from IPython.core.display import Image
Image(filename="show-to-display.jpg")
```

# Until is Done

```python
from SimpleCV import Display, Image
import time
display = Display()
Image("logo").save(display)
print "I launched a window"
# This while loop will keep looping until the window is closed (via a right-click)
while not display.isDone(): 
  time.sleep(0.1)
print "You closed the window"
Display().quit()
```

# Painting circles

```python
from SimpleCV import Display, Color, Image

winsize = (640,480)
display = Display(winsize)

img = Image(winsize)
img.save(display)

while not display.isDone():
  if display.mouseLeft:
    img.dl().circle((display.mouseX, display.mouseY), 4,
    Color.WHITE, filled=True)
    img.save(display)
    img.save("painting.png")
```

```python
from IPython.core.display import Image
Image(filename="painting.png")
```

# Time-lapse Photography

```python
from SimpleCV import Camera, Image
import time

cam = Camera()

# Set the number of frames to capture
numFrames = 3

# Loop until we reach the limit set in numFrames
for x in range(0, numFrames):
  img = cam.getImage()
  
  filepath = "image-" + str(x) + ".jpg"
  img.save(filepath)
  print "Saved image to: " + filepath
  
  time.sleep(2)
```

# Photobooth

```python
from SimpleCV import Camera, Display, Color
import time
# Initialize the camera
cam = Camera()
# Initialize the display
display = Display()
# Take an initial picture
img = cam.getImage()
# Write a message on the image
img.drawText("Left click to save a photo.",
color=Color().getRandom())
# Show the image on the display
img.save(display)
time.sleep(5)
counter = 0
while not display.isDone():
  # Update the display with the latest image
  img = cam.getImage()
  img.save(display)
  if display.mouseLeft:
    # Save image to the current directory
    img.save("photobooth" + str(counter) + ".jpg")
    img.drawText("Photo saved.", color=Color().getRandom())
    img.save(display)
    time.sleep(3)
    counter = counter + 1
```
